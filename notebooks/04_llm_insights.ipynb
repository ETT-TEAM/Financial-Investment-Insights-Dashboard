# 04 — LLM-Inspired Insight Generation

This notebook demonstrates the **rule-based Insight Engine** that generates
human-readable, LLM-style financial insights from our processed data.

The engine analyses:
- Stock risk-return tradeoffs
- Highest/lowest performers
- Mutual fund category performance
- Long-term consistency patterns

---

```python
import pandas as pd
import numpy as np
import os, sys
import warnings
warnings.filterwarnings('ignore')

# Add src/ to path
sys.path.insert(0, os.path.join('..', 'src'))

from data_processing import load_and_clean_all_stocks, load_raw_mutual_funds, clean_mutual_funds
from metrics import compute_stock_metrics, risk_return_comparison, compute_mf_metrics, top_mutual_funds
from insight_engine import InsightEngine

RAW_DIR = os.path.join('..', 'data', 'raw')
TICKERS = ['AAPL', 'MSFT', 'GOOG']
```

```python
# Load and clean data
stocks = load_and_clean_all_stocks(TICKERS, RAW_DIR)
mf = clean_mutual_funds(load_raw_mutual_funds(RAW_DIR))

# Compute metrics
stock_metrics = compute_stock_metrics(stocks)
print('Stock Metrics:')
display(stock_metrics)

rr = risk_return_comparison(stock_metrics)
print(f'\nHighest return: {rr["highest_return"]}')
print(f'Lowest risk : {rr["lowest_risk"]}')
print(f'Best Sharpe : {rr["best_sharpe"]}')
```

## 1. Initialize the Insight Engine

```python
# Create the insight engine
engine = InsightEngine(
 stock_metrics=stock_metrics,
 mf_df=mf,
 stock_data=stocks
)

print(' InsightEngine initialized successfully')
print(f' Stocks analysed : {len(stock_metrics)} tickers')
print(f' Mutual funds  : {len(mf)} schemes')
```

## 2. Generate Individual Insights

```python
# Market overview
print(engine.market_overview())
```

```python
# Highest return stock
print(engine.highest_return_stock())
```

```python
# Lowest risk stock
print(engine.lowest_risk_stock())
```

```python
# Risk-return tradeoff
print(engine.risk_return_tradeoff())
```

```python
# Cumulative performance
print(engine.cumulative_performance())
```

```python
# Top mutual fund performers
print(engine.mf_top_performers(5))
```

```python
# Category analysis
print(engine.mf_category_analysis())
```

```python
# Mutual fund risk-return insight
print(engine.mf_risk_return_insight())
```

```python
# Long-term consistency
print(engine.mf_long_term_consistency())
```

## 3. Full AI Insight Report

```python
# Generate the complete report
report = engine.generate_report()
print(report)
```

## 4. Dictionary Output (for Dashboard Integration)

```python
# Generate insights as a dictionary
insights_dict = engine.generate_dict()

for key, value in insights_dict.items():
 print(f'\n--- {key} ---')
 print(value)
```

## 5. Export Insights for Dashboard

```python
# Save insights as a text file for Power BI text panel
out_dir = os.path.join('..', 'data', 'processed')
os.makedirs(out_dir, exist_ok=True)

# Save full report
report_path = os.path.join(out_dir, 'ai_insights_report.txt')
with open(report_path, 'w', encoding='utf-8') as f:
 f.write(report)
print(f'✓ Report saved → {report_path}')

# Save as CSV for dashboard integration
insights_df = pd.DataFrame([
 {'category': k.replace('_', ' ').title(), 'insight': v}
 for k, v in insights_dict.items()
])
csv_path = os.path.join(out_dir, 'ai_insights.csv')
insights_df.to_csv(csv_path, index=False)
print(f'✓ Insights CSV saved → {csv_path}')

display(insights_df)
```
