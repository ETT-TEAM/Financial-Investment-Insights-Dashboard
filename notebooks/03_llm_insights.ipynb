```python
# Load mutual fund dataset
mutual = pd.read_csv("/comprehensive_mutual_funds_data.csv")

# Clean column names
mutual.columns = mutual.columns.str.strip().str.lower()
```

```python
volatility = {
    "Apple": aapl_overlap['daily_return'].std(),
    "Microsoft": msft_overlap['daily_return'].std(),
    "Google": goog_overlap['daily_return'].std()
}

returns = {
    "Apple": aapl_overlap['daily_return'].mean(),
    "Microsoft": msft_overlap['daily_return'].mean(),
    "Google": goog_overlap['daily_return'].mean()
}
```

```python
for stock in returns:
    print(f"{stock}:")

    if returns[stock] > 0:
        print("- Positive average return")
    else:
        print("- Negative average return")

    print(f"- Volatility: {volatility[stock]:.4f}")
    print()
```

```python
best_return = max(returns, key=returns.get)
lowest_risk = min(volatility, key=volatility.get)

print(f"{best_return} has the highest returns.")
print(f"{lowest_risk} is the least risky stock.")
```

```python
high_return_funds = mutual[mutual['returns_5yr'] > 15]

print("Funds with high returns (>15%):")
print(high_return_funds[['scheme_name','returns_5yr']].head())
```

```python
summary = f"""
From the analysis:

- {best_return} shows the highest average return.
- {lowest_risk} has the lowest volatility, making it less risky.
- Stocks with higher volatility tend to provide higher returns.

In mutual funds:
- High return funds generally have higher risk (standard deviation).
- Most funds lie in moderate risk-return range.

Conclusion:
There is a clear trade-off between risk and return.
"""

print(summary)
```
