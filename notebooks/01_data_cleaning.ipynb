```python
import pandas as pd
import numpy as np
```

```python
aapl = pd.read_csv("AAPL_raw.csv")
msft = pd.read_csv("MSFT_raw.csv")
goog = pd.read_csv("GOOG_raw.csv") # Read with correct headers
goog = goog.drop(index=0).reset_index(drop=True) # Drop the problematic row at index 0 and reset index

mutual = pd.read_csv("comprehensive_mutual_funds_data.csv")
```

```python
aapl.head()
```

```python
aapl['Date'] = pd.to_datetime(aapl['Date'])
msft['Date'] = pd.to_datetime(msft['Date'])
goog['date'] = pd.to_datetime(goog['date'])
```

```python
print(aapl.columns)
```

```python
aapl = aapl.sort_values('Date')
msft = msft.sort_values('Date')
goog = goog.sort_values('date')
```

```python
aapl[' Close/Last'] = aapl[' Close/Last'].astype(str).str.replace('$', '', regex=False).astype(float)
aapl['Daily_Return'] = aapl[' Close/Last'].pct_change()

msft['Close'] = msft['Close'].astype(str).str.replace('$', '', regex=False).astype(float)
msft['Daily_Return'] = msft['Close'].pct_change()

goog['close'] = goog['close'].astype(str).str.replace('$', '', regex=False).astype(float)
goog['Daily_Return'] = goog['close'].pct_change()
```

```python
aapl.head()
```

```python
aapl_volatility = aapl['Daily_Return'].std()
msft_volatility = msft['Daily_Return'].std()
goog_volatility = goog['Daily_Return'].std()

print("Apple Volatility:", aapl_volatility)
print("Microsoft Volatility:", msft_volatility)
print("Google Volatility:", goog_volatility)
```

```python
aapl.to_csv("AAPL_clean.csv", index=False)
msft.to_csv("MSFT_clean.csv", index=False)
goog.to_csv("GOOG_clean.csv", index=False)
```
