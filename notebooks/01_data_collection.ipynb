# 01 — Data Collection & Overview

This notebook loads the raw financial datasets and provides an initial overview of the data.

**Datasets:**
- Stock data: Apple (AAPL), Microsoft (MSFT), Google (GOOG)
- Comprehensive Mutual Funds dataset (814 funds)

---

```python
# Standard imports
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

# Configuration
RAW_DIR = os.path.join('..', 'data', 'raw')
print('Raw data directory:', os.path.abspath(RAW_DIR))
print('Files available:', os.listdir(RAW_DIR))
```

## 1. Stock Data — Loading Raw Files

```python
# Load raw stock CSVs
tickers = ['AAPL', 'MSFT', 'GOOG']
stock_raw = {}

for ticker in tickers:
 path = os.path.join(RAW_DIR, f'{ticker}_raw.csv')
 stock_raw[ticker] = pd.read_csv(path)
 print(f'\n{"=" * 50}')
 print(f' {ticker} — Raw Data Overview')
 print(f'{"=" * 50}')
 print(f' Shape : {stock_raw[ticker].shape}')
 print(f' Columns : {stock_raw[ticker].columns.tolist()}')
 print(f' Dtypes :')
 print(stock_raw[ticker].dtypes.to_string())
 print(f'\n First 3 rows:')
 display(stock_raw[ticker].head(3))
```

## 2. Stock Data — Initial Quality Check

```python
# Check for missing values and data quality issues
for ticker in tickers:
 df = stock_raw[ticker]
 print(f'\n--- {ticker} Data Quality ---')
 print(f'Missing values:\n{df.isnull().sum()}')
 print(f'Duplicated rows: {df.duplicated().sum()}')
 print(f'Date range: {df.iloc[:, 0].min()} to {df.iloc[:, 0].max()}')
```

## 3. Mutual Funds Data — Loading

```python
# Load mutual fund data
mf_raw = pd.read_csv(os.path.join(RAW_DIR, 'comprehensive_mutual_funds_data.csv'))

print('=' * 50)
print(' Mutual Funds — Raw Data Overview')
print('=' * 50)
print(f'Shape : {mf_raw.shape}')
print(f'Columns : {mf_raw.columns.tolist()}')
print(f'\nData Types:')
print(mf_raw.dtypes)

display(mf_raw.head(5))
```

```python
# Mutual fund quality check
print('Missing values per column:')
print(mf_raw.isnull().sum())
print(f'\nTotal missing: {mf_raw.isnull().sum().sum()}')
print(f'Duplicated rows: {mf_raw.duplicated().sum()}')
print(f'\nUnique categories: {mf_raw["category"].nunique()}')
print(mf_raw['category'].value_counts())
```

## 4. Summary Statistics

```python
# Statistical summary of stock data
for ticker in tickers:
 print(f'\n--- {ticker} Descriptive Statistics ---')
 display(stock_raw[ticker].describe())
```

```python
# Statistical summary of mutual fund data
print('--- Mutual Funds Descriptive Statistics ---')
display(mf_raw.describe())
```

## 5. Key Observations

### Stock Data
- Each stock file contains ~2500 trading days of data
- Columns: Date, Close/Last, Volume, Open, High, Low
- Price columns contain '$' symbols that need removal
- Volume has comma separators
- Date needs conversion to datetime format

### Mutual Fund Data
- 814 mutual fund schemes across multiple categories
- 20 features including returns (1yr, 3yr, 5yr), risk metrics (sd, beta, sharpe)
- Some missing values in numeric columns
- Categories need standardization
